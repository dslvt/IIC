```python
model_name = ['orig', 'v3', 'v4']
cos_sim = [0.32895532, 0.31928322, 0.5209617]
bleu = [0.0037887643258351193, 0.007913815471841354, 0.08682854120861037]
dataset_size = [0, 30000, 110000]
image_to_text_mean_rank = [78.8552, 105.6876, 35.8462, ]
image_to_text_R1 = [0.4352, 0.4217, 0.5070]
image_to_text_R10 = [0.7760, 0.7482, 0.8368]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, cos_sim)

ax.set_ylabel('cosine similarity')
ax.set_xlabel('model name')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, bleu)

ax.set_ylabel('BLEU')
ax.set_xlabel('model name')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, dataset_size)

ax.set_ylabel('Dataset size')
ax.set_xlabel('model name')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, image_to_text_mean_rank)

ax.set_ylabel('Image to text mean rank')
ax.set_xlabel('model name')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, image_to_text_R1)

ax.set_ylabel('Image to text R@1')
ax.set_xlabel('model name')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(model_name, image_to_text_R10)

ax.set_ylabel('Image to text R@10')
ax.set_xlabel('model name')

plt.show()
```
